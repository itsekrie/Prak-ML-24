Import library

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import load_iris
```

Mengolah data bunga Iris dengan memilih dua fitur spesifik, yaitu lebar sepal dan panjang petal, kemudian menyimpan hasilnya dalam array baru bernama `X`.

```python
iris = load_iris()
features = iris.data.T
X = np.array(list(zip(features[1], features[2])))
```

Membuat visualisasi data bunga Iris yang menunjukkan hubungan antara lebar sepal dan panjang petal, misalnya menggunakan scatter plot untuk menggambarkan pola atau distribusi data.

```python
plt.scatter(features[1], features[2])
plt.show()
```

Melakukan clustering pada data X menggunakan algoritma K-Means dengan jumlah cluster sebanyak 3, kemudian mencetak koordinat pusat-pusat cluster yang dihasilkan.

```python
# Menentukan jumlah cluster
kmeans = KMeans(n_clusters=3)
# Fitting input data
kmeans = kmeans.fit(X)
# Mendapatkan cluster labels
labels = kmeans.predict(X)
# Mendapatkan nilai centroid
C = kmeans.cluster_centers_
# Mencetak nilai centroid
print(C)
```

Membuat visualisasi hasil clustering yang telah dilakukan sebelumnya untuk mempermudah pengamatan pola pengelompokan data berdasarkan hasil clustering. Setiap cluster akan diberi warna berbeda untuk membedakan kelompoknya.

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=7)
plt.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
```

Kode di atas bertujuan untuk menentukan jumlah cluster yang optimal untuk data `X` menggunakan algoritma K-Means. Proses ini dilakukan dengan mencoba berbagai nilai `k` (jumlah cluster) mulai dari 1 hingga 9, untuk kemudian mengevaluasi hasil clustering dan memilih nilai `k` yang memberikan hasil terbaik.

```python
for k in range (1,10):
  # Menentukan jumlah cluster
  kmeans = KMeans(n_clusters=k, random_state=1)
  # Fitting input data
  kmeans = kmeans.fit(X)
  # Mendapatkan cluster labels
  labels = kmeans.predict(X)
  # Menghitung jumlahan jarak anrara setiap sampel dengan cluster centroidnya (SSE)
  inertia = kmeans.inertia_
  print("k:",k, "cost:",inertia)
```

Silhouette_score(X, labels) adalah sebuah fungsi dalam Python yang digunakan untuk menghitung skor siluet pada hasil clustering

```python
silhouette_score(X, labels)
```

Menemukan jumlah cluster yang optimal untuk data Anda menggunakan algoritma KMeans. KMeans adalah algoritma pengelompokan tak terawasi (unsupervised), yang berarti algoritma ini berusaha mengelompokkan data ke dalam cluster secara otomatis tanpa menggunakan label kelas yang telah ditentukan sebelumnya.

```python
X = np.array(list(zip(features[1], features[2]))).reshape(len(features[1]), 2)
SSE = []
K = range(1, 10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    SSE.append(kmeanModel.inertia_)
```

Dengan membuat plot menggunakan kode di atas, kita dapat secara visual melihat hubungan antara jumlah cluster dan nilai SSE (Sum of Squared Errors). Titik siku pada plot, yang menunjukkan perubahan tajam dalam penurunan SSE, akan membantu kita menentukan jumlah cluster yang paling optimal untuk data kita. Titik siku ini memberikan indikasi jumlah cluster yang memberikan hasil clustering yang lebih baik dengan mengurangi kesalahan dalam pengelompokan data.

```python
plt.plot(K, SSE, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()
```

Kode tersebut melakukan analisis pengelompokan data tanpa pengawasan (unsupervised) menggunakan algoritma K-Means dari library Scikit-learn. Berikut langkah-langkah yang dilakukan:

Baris pertama membuat model K-Means dengan jumlah cluster yang diinginkan (3 cluster) dan menetapkan seed untuk menghasilkan hasil yang konsisten.
Baris kedua memprediksi label cluster untuk setiap data menggunakan metode fit_predict(), hasilnya disimpan dalam variabel y_pred.
Kemudian, dilakukan visualisasi hasil pengelompokan. Fungsi plot() digunakan untuk membuat plot kosong, kemudian titik data diplot berdasarkan fitur kedua dan ketiga (features[1] dan features[2]).
Warna titik ditentukan berdasarkan label cluster menggunakan label_color yang dipetakan dari LABEL_COLOR_MAP.
Terakhir, judul plot ("K=3") ditambahkan dan keseluruhan plot ditampilkan menggunakan plt.show().
Dengan langkah-langkah ini, kita dapat memvisualisasikan bagaimana data dikelompokkan dalam 3 cluster.








```python
y_pred = KMeans(n_clusters=3, random_state=0).fit_predict(X)
plt.plot()
LABEL_COLOR_MAP = {0: 'r', 1: 'g', 2: 'b'}
label_color = [LABEL_COLOR_MAP[l] for l in y_pred]
plt.scatter(features[1], features[2], c=label_color)
plt.title('K=3')
plt.show()
```
